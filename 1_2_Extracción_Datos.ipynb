```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

CARGA DE ARCHIVO XLSX

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturación.xlsx')
df 
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
#1. CVE_CLPV
filtro1 =df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1.to_csv("Filtro 1.CVE_CLPV")
filtro1
```

```python
#2. CVE_VEND
filtro2 =df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtro2.to_csv("Filtro 2.CVE_VEND")
filtro2
```

```python
#3. FECHA_ENT
filtro3 = df[df["FECHA_ENT"] == "28/02/2022 23:59:59.000"]
filtro3.to_csv("Filtro 3.FECHA_ENT")
filtro3
```

```python
#4. CAN_TOT
filtro4 =df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4.to_csv("Filtro 4.CAN_TOT")
filtro4
```

```python
#5. COLUMNAS
filtro5 =df.iloc[ : , [0,6,7,9]] 
filtro5.to_csv("Filtro 5.COLUMNAS")
filtro5
```

```python
#6. FILA
filtro6 =df.iloc[7001:7100,: ]
filtro6.to_csv("Filtro 6.CAN_TOT")
filtro6
```

```python
#7. INDEX
df1= pd.read_excel('Datos de Facturación.xlsx')
df1 = df1.set_index('CVE_VEND')
filtro4 =df1.loc[[1,2], ["FECHAELAB"]]
filtro4.to_csv("Filtro 7.INDEX")
filtro4
```
