```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

CARGA DE ARCHIVO XLSX

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
#df 
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto 
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1.to_csv("Filtro 1 por objeto")
filtro1
```

```python
#Filtro por filas
filtro2=df.iloc[ [5, 25, 50] , : ]  #Filas no consecutivas 
filtro2.to_csv("Filtro 2 por filas")
filtro2
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , 1:4]       #Columnas consecutivas 
filtro3.to_csv("Filtro 3 por columnas") 
filtro3
```

```python
#Filtro de filas vs columnas 
#Paso 1: Leemos el contenido del archivo
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=1)
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
filtro4=df1.loc[["2022-01-05","2022-12-23"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4.to_csv("Filtro 4 filas vs columnas") 
filtro4
```

```python
#Filtro por cabecera
filtro5=df.head(5)
filtro5.to_csv("Filtro 5 por cabecera") 
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
filtro6=df[df["MARGEN_TOTAL_CALCULADO"] > 1000]
filtro6.to_csv("Filtro 6 por comparación") 
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["MARGEN_TOTAL_CALCULADO"] > 500) & (df["FECHA_DOC"] == "2022-12-23")]
filtro7.to_csv("Filtro 7 Y") 
filtro7
```

```python
#Filtro o
filtro8=df[(df["MARGEN_TOTAL_CALCULADO"] > 700) | (df["FECHA_DOC"] == "2022-12-23")]
filtro8.to_csv("Filtro 8 o") 
filtro8
```

```python
#Filtro not
filtro9=df[~(df["MARGEN_TOTAL_CALCULADO"] <= 1000) & ~(df["FECHA_DOC"] == "2022-12-23")]
filtro9.to_csv("Filtro 9 not") 
filtro9
```
